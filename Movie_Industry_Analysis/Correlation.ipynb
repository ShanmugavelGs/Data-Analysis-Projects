```python
# Import Libraries

import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import figure
import re

%matplotlib inline
plt.style.use('ggplot')
matplotlib.rcParams['figure.figsize'] = (8,6)
```

```python
# Read in the data

df = pd.read_csv('movies.csv')
```

```python
# Let's look at the data

df.head()
```

```python
df.shape
```

```python
# Let's see if there is any missing data

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print(f'{col} - {pct_missing:.2f}%')
```

```python
# Drop the missing values

df = df.dropna()
```

```python
# data types of our columns

df.dtypes
```

```python
# Change data types of our columns

df['budget'] = pd.to_numeric(df['budget'], errors='coerce').fillna(0).astype('int64')
df['gross'] = pd.to_numeric(df['gross'], errors='coerce').fillna(0).astype('int64')
```

```python
df.head(10)
```

```python
# Extract year from the string

df['correctyear'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
df
```

```python
pd.set_option('display.max_rows',100)
```

```python
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# Drop any duplicates

df.drop_duplicates()
```

```python
# Check for Correlations

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget')
plt.ylabel('Gross Revenue')
plt.show()
```

```python
# plot using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
# Let's start looking at correlation

numeric_df = df.select_dtypes(include=['number'])
numeric_df.corr(method='pearson')
```

```python
# High correlation between budget and gross
```

```python
correlation_matrix = numeric_df.corr(method='pearson')
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```
